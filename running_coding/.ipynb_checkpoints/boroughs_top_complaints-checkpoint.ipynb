```python

import requests
import pandas as pd
import numpy as np
import csv
# from . import support  relative path example
from . import boroughs_array
from . import main_data_dataframe # maybe don't need to be recreated this way



```

```python
# helper variables

popular_complaints = pd.Series(main_data_dataframe.groupby('Complaint Type').size().sort_values(ascending=False))
top_ten_pop_complaints = pd.Series(popular_complaints[:10])
top_ten_pop_complaints_keys = top_ten_pop_complaints.keys()
complaints_tops_names = top_ten_pop_complaints.keys()
```

```python
def define_base_obj_to_use(): # sets obj to use for most popular complaints. includes zero
    starter_complaints_obj = {}
    for x in complaints_tops_names:
        starter_complaints_obj[x] = 0
    return starter_complaints_obj


def new_group_boroughs_by_zips():

    borough_list = ["BROOKLYN", "QUEENS", "BRONX", "STATEN ISLAND", "MANHATTAN"]
    borough_zips = {
    "BROOKLYN" : [],
    "QUEENS": [],
    "BRONX": [],
    "STATEN ISLAND": [],
    "MANHATTAN": []
}
    
    for zipcode in bronxtable['Zip']: 
        (borough_zips['BRONX'].append(float(zipcode)))
    for zipcode in manhattan_table['Zip']:
        borough_zips['MANHATTAN'].append(float(zipcode))
    for zipcode in queens_table['Zip']:
        borough_zips['QUEENS'].append(float(zipcode))
    for zipcode in staten_island_table['Zip']:
        borough_zips['STATEN ISLAND'].append(float(zipcode))
    for zipcode in brooklyn_table['Zip']:
        borough_zips['BROOKLYN'].append(float(zipcode))
    return(borough_zips)



def create_borough_complaints_obj():
    empty_complaints_obj = define_base_obj_to_use()
    obj = {}
    for x in boroughs_array:
        obj[x] = empty_complaints_obj
    return obj

```

```python
bronxtable = main_data_dataframe.loc[main_data_dataframe.Borough == 'BRONX']
manhattan_table = main_data_dataframe.loc[main_data_dataframe.Borough == 'MANHATTAN']
queens_table = main_data_dataframe.loc[main_data_dataframe.Borough == 'QUEENS']
staten_island_table = main_data_dataframe.loc[main_data_dataframe.Borough == 'STATEN ISLAND']
brooklyn_table = main_data_dataframe.loc[main_data_dataframe.Borough == 'BROOKLYN']
bronx_complaints =  bronxtable["Complaint Type"]

### Row Values of Complaint Types. made for convenience 


manhattan_complaints = manhattan_table["Complaint Type"]
queens_complaints = queens_table["Complaint Type"]
SI_complaints = staten_island_table["Complaint Type"]
brooklyn_complaints = brooklyn_table["Complaint Type"]
bronx_complaints =  bronxtable["Complaint Type"]

total_table = manhattan_complaints.append([queens_complaints, SI_complaints, brooklyn_complaints, bronx_complaints])
```

```python
def nyc_complainers(): 
    borough_obj = {'BRONX': {}, 'QUEENS': {}, 'MANHATTAN' : {}, 'STATEN ISLAND': {}, 'BROOKLYN': {} }
    bronx_complainers_dict = create_borough_complaints_obj()['BRONX']

    for complaint in bronx_complaints:  
        if complaint not in complaints_tops_names:
            pass
        elif (complaint in bronx_complainers_dict and complaint in complaints_tops_names):
            bronx_complainers_dict[complaint] +=1
        else:
            bronx_complainers_dict[complaint] = 1
    
    queens_complainers_dict = create_borough_complaints_obj()['QUEENS']
    for complaint in queens_complaints: 
        if complaint not in complaints_tops_names:
            pass
        elif (complaint in queens_complainers_dict and complaint in complaints_tops_names):
            queens_complainers_dict[complaint] +=1
        else:
            queens_complainers_dict[complaint] = 1

    manhattan_complainers_dict = create_borough_complaints_obj()['MANHATTAN']
    for complaint in manhattan_complaints: 
        if complaint not in complaints_tops_names:
            pass
        elif (complaint in manhattan_complainers_dict and complaint in complaints_tops_names):
            manhattan_complainers_dict[complaint] +=1
        else:
            manhattan_complainers_dict[complaint] = 1

    si_complainers_dict = create_borough_complaints_obj()['STATEN ISLAND']
    for complaint in SI_complaints: 
        if complaint not in complaints_tops_names:
            pass
        elif (complaint in si_complainers_dict and complaint in complaints_tops_names):
            si_complainers_dict[complaint] +=1
        else:
            si_complainers_dict[complaint] = 1

    brooklyn_complainers_dict = create_borough_complaints_obj()['BROOKLYN']
    for complaint in brooklyn_complaints: 
        if complaint not in complaints_tops_names:
            pass
        elif (complaint in brooklyn_complainers_dict and complaint in complaints_tops_names):
            brooklyn_complainers_dict[complaint] +=1
        else:
            brooklyn_complainers_dict[complaint] = 1
    borough_obj['BRONX'] = bronx_complainers_dict
    borough_obj['BROOKLYN'] = brooklyn_complainers_dict
    borough_obj['STATEN ISLAND'] = si_complainers_dict
    borough_obj['QUEENS'] = queens_complainers_dict
    borough_obj['MANHATTAN'] = manhattan_complainers_dict
    return borough_obj
```

```python

bronx_complainers_var = nyc_complainers()['BRONX']
queens_complainers_var = nyc_complainers()['QUEENS']
manhattan_complainers_var = nyc_complainers()['MANHATTAN']
si_complainers_var = nyc_complainers()['STATEN ISLAND']
brooklyn_complainers_var = nyc_complainers()['BROOKLYN']
```
