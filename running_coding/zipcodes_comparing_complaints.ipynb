```python
def define_zipcode_tables(): 
    obj = {}
    for x in range(10): 
        obj[x] = main_data_dataframe.loc[main_data_dataframe.Zip == pop_zip_array[x]]
        # obj[pop_zip_array[x]] = main_data_dataframe.loc[main_data_dataframe.Zip == float(pop_zip_array[x])]
    return obj
```
